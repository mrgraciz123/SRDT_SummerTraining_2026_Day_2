<h2 align='center' style='color:blue'>Exercise: Linear Regression Model for Canada's Per Capita Income Prediction</h2>

You are given `canada_per_capita_income.csv` which contains the historical per capita income of Canadian citizens. You need to build a regression model and predict the per capita income of Canadian citizens in the year **2020**.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the dataset
df = pd.read_csv("canada_per_capita_income.csv")
df.head(10)
```

Let's examine the columns and check the first few details of the dataset.

```python
df.info()
df.describe()
```

### Plot a scatter plot to visualize the trend of income over the years

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='#2c3e50', marker='o')
plt.title("Canada's Per Capita Income over Years", fontsize=14, fontweight='bold')
plt.xlabel("Year", fontsize=12)
plt.ylabel("Per Capita Income (US$)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

### Build and Train the Linear Regression Model

```python
# Independent variable (X) needs to be 2D
X = df[['year']]
# Dependent variable (y)
y = df['per capita income (US$)']

# Create regression object
reg = LinearRegression()
reg.fit(X, y)

print("Coefficient (Slope m):", reg.coef_[0])
print("Intercept (c):", reg.intercept_)
print(f"Equation: Income = {reg.coef_[0]:.4f} * Year + ({reg.intercept_:.4f})")
```

### Predict Canada's Per Capita Income in Year 2020

```python
year_to_predict = 2020
# Wrap the target year in a DataFrame to avoid feature name warnings
predict_df = pd.DataFrame({'year': [year_to_predict]})
predicted_income = reg.predict(predict_df)[0]

print(f"Predicted Per Capita Income of Canadian citizens in year {year_to_predict}: ${predicted_income:,.2f}")
```

### Plot the linear regression line on the scatter plot along with the predicted 2020 point

```python
plt.scatter(df['year'], df['per capita income (US$)'], color='#2c3e50', marker='o', label='Actual Income')
plt.plot(df['year'], reg.predict(X), color='#e74c3c', linewidth=2, label='Regression Line')

# Highlight the prediction point for 2020
plt.scatter([year_to_predict], [predicted_income], color='#27ae60', marker='*', s=200, zorder=5, 
            label=f'Predicted 2020: ${predicted_income:,.2f}')

plt.title("Canada's Per Capita Income Regression Model", fontsize=14, fontweight='bold', pad=15)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Per Capita Income (US$)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```
