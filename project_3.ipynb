## Project 3: Predict Canada’s per capita income in year 2020

```python
import pandas as pd

df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

Next, we will train a linear regression model to predict the per capita income based on the year. We'll then use this model to predict the income for the year 2020.

```python
from sklearn import linear_model

# Create a linear regression object
reg = linear_model.LinearRegression()

# Train the model using the 'year' as the feature and 'per capita income (US$)' as the target
reg.fit(df[['year']], df['per capita income (US$)'])

# Predict the per capita income for the year 2020
income_2020 = reg.predict([[2020]])

print(f"Predicted per capita income for Canada in 2020: {income_2020[0]:.2f} US$")
```
